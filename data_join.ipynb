```python
import numpy as np
import pandas as pd

users_df = pd.read_csv("cleaned_datasets/users_details_dataset_cleaned.csv")
anime_df = pd.read_csv("cleaned_datasets/anime_dataset_cleaned.csv")
user_score_df = pd.read_csv("cleaned_datasets/user_scores_cleaned.csv")
anime_df.info()
```

```python
# Keep users that are in both CSVs
joined_rating_dataset = pd.merge(user_score_df, users_df, left_on=['user_id', 'Username'], right_on=['Mal ID', 'Username'], how='inner')
joined_rating_dataset = joined_rating_dataset[["user_id", "anime_id", "rating", "Gender", "Location", "Birthday_Date","Joined_Date","Age_Join", "Episodes Watched"]]
joined_rating_dataset.info()
```

```python
type(joined_rating_dataset["Birthday_Date"][20])
joined_rating_dataset["Age"] = ((pd.to_datetime('Jan 01, 2024', format='%b %d, %Y') - pd.to_datetime(joined_rating_dataset["Birthday_Date"], format='%Y-%m-%d')) / np.timedelta64(52, 'W'))
joined_rating_dataset = joined_rating_dataset.round()
joined_rating_dataset.head()
```

```python
anime_df = anime_df[["anime_id", "Name", "Genres", "Type", "Start Date", "End Date", "Studios", "Source", "Rank", "Episodes", "Episodes_Norm"]]
joined_rating_dataset = pd.merge(joined_rating_dataset, anime_df, left_on=['anime_id'], right_on=['anime_id'], how='inner')
joined_rating_dataset.info()

```

```python
joined_rating_dataset["Episodes"] = joined_rating_dataset["Episodes"].round()
joined_rating_dataset["Age_Join"] = joined_rating_dataset["Age_Join"].round()
joined_rating_dataset.head()
```
