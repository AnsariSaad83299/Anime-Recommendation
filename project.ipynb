```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/Users/ramachandrank/Documents/MS/Sem 1/DIC/project1/users-details-2023.csv")
df.info()
```

```python
df_2 = df[~df["Birthday"].isna()]
df_2 = df_2[~df_2["Joined"].isna()]
df_2
```

```python

# Remove time stamp from date string
df_2["Birthday_Date"] = df_2["Birthday"].str.slice(0,10)
df_2["Joined_Date"] = df_2["Joined"].str.slice(0,10)

# Convert string to date time object
df_2["Birthday_Date"] = pd.to_datetime(df_2["Birthday_Date"], format="%Y-%m-%d", errors = 'coerce')
df_2["Joined_Date"] = pd.to_datetime(df_2["Joined_Date"], format="%Y-%m-%d", errors = 'coerce')

# Remove entries where joining date is before birthdate
df_2 = df_2[df_2["Birthday_Date"] < df_2["Joined_Date"]]

# Remove entries where birthday is very very old
df_2 = df_2[df_2["Birthday_Date"] > pd.to_datetime("1950-01-01",format="%Y-%m-%d")]

# Remove entries where age at joining is less than 5 years
df_2["Age_Join"] = (df_2["Joined_Date"] - df_2["Birthday_Date"]) / np.timedelta64(52, 'W')
df_2 = df_2[df_2["Age_Join"]>5]

df_2[["Birthday_Date","Joined_Date", "Age_Join"]].sort_values(by="Age_Join")
```
